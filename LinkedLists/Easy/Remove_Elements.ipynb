#### Remove Elements from Linked List (Easy) [TIP102-2-1]

```python
class ListNode:
    def __init__(self, val, next=None):
        self.val = val
        self.next = next
```

```python
def remove_elements(head, target):
    temp_head = ListNode(None, next=head)
    
    current = temp_head
    while current.next:
        if current.next.val == target:
            current.next = current.next.next
        
        else:
            current = current.next
    
    return temp_head.next 
```
