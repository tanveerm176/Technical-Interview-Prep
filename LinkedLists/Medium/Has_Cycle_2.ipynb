### Has Cycle 2 (Meidum) [TIP102-2-1]
- LeetCode: https://leetcode.com/problems/linked-list-cycle-ii/
- Video Solution: https://www.youtube.com/watch?v=95ZfuoSAUPI



#### Understand
    1. Can there be more than one cycle?
    2. Can the list be empty?
    3. Will there always be a cycle?

##### Cases
    HAPPY CASE
    Input: head = [3,2,0,-4], pos = 1
    Output: 1
    Explanation: There is a cycle in the linked list, where the tail connects to the 1st node (0-indexed).

    Input: head = [1,2], pos = 0
    Output: 0
    Explanation: There is a cycle in the linked list, where the tail connects to the 0th node.

    EDGE CASE
    Input: head = [1], pos = -1
    Output: -1
    Explanation: There is no cycle in the linked list.

##### Drawings







#### Match 
    1. Using two pointers like in Has_Cycle 1 we can determine the node from which the cycle starts

#### Plan with Fake Code
    if empty list return 
    init slow and fast pointers 
    while not at last node and not at None:
        move slow by 1
        move fast by 2

        if fast == slow:
            create new pointer at head
            while new pointer != slow:
                move slow by 1
                move new pointer by 1
            
            return new pointer

    return none if all nodes processed and no cycle found

> this works because of a mathematical understanding of how cicrles can intersect with lines at a tangent

#### Implement 

```python
class ListNode:
    def __init__(self, val, next=None):
        self.val = val
        self.next = next
```

```python
def has_cycle_2(head: ListNode):
    # init slow and fast pointers
    slow = head
    fast = head

    # while not at last node and not at None:
    while fast and fast.next:
        # move slow by 1
        # move fast by 2
        slow = slow.next
        fast = fast.next.next

        # if fast == slow:
        if fast == slow:
            # create new pointer at head
            temp_ptr = head

            # while new pointer != slow:
            while temp_ptr != slow:
                # move slow by 1
                # move new pointer by 1
                slow = slow.next
                temp_ptr = temp_ptr.next

            # return new pointer
            return slow

    # processed entire list no cycles found
    return None
```
