### Max Depth Binary Tree
https://leetcode.com/problems/maximum-depth-of-binary-tree/description/



```python
class Node:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def list_to_bin_tree(l):
    if not l:
        return None

    root = Node(l.pop(0))
    queue = [root]

    while queue:
        node = queue.pop(0)

        if l:
            left_val = l.pop(0)
            if left_val != None:
                node.left = Node(left_val)
                queue.append(node.left)

        if l:
            right_val = l.pop(0)
            if right_val != None:
                node.right = Node(right_val)
                queue.append(node.right)

    return root


def bin_tree_to_list(root):
    l = []
    if not root:
        return l
    queue = [root]

    while queue:
        node = queue.pop(0)
        l.append(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

    return l

# recursively find the max of the left and right subtrees
# recursive DFS
# Time: 0(n), Space: 0(n) worst case for unbalanced tree
def maxDepth_DFS_recursive(root):
    if not root: return 0
    
    """ 
    using DFS recursively find the max of the left and right subtrees until the leaf node
    adding 1 to each to account for the root node since it exists
    """
    return 1 + max(maxDepth_DFS_recursive(root.left), maxDepth_DFS_recursive(root.right))




```

#### **BFS level order maxDepth solution**


```python
# Use level order bfs to find max depth, count the number of levels 
# Does not use recursion
# Same time and space complexity
from collections import deque
def maxDepth_BFS(root):
    if not root: return 0
    
    # level counter var
    level_count = 0
    # Init queue with root 
    q = deque([root])
    
    while q is not None:
        
        # iterate over current nodes in q and remove the first node, 
        # adding it's children to the q
        # inc the level counter after child nodes have been added
        for _ in range(len(q)):
            
            curr_node = q.popleft()
            
            if curr_node.left:
                q.append(curr_node.left)
            
            if curr_node.right:
                q.append(curr_node.right)

        level_count += 1
        
    return level_count
```
