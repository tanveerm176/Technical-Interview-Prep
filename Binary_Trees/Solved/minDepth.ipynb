### Minx Depth Binary Tree

```python
def min_depth(root):
    if not(root): return 0
    
    if root.left and root.right:
        return min(min_depth(root.left), min_depth(root.right)) + 1
    elif root.right:
        return min_depth(root.right) + 1
    elif root.left:
        return min_depth(root.left) + 1
    
    else:
        return 1
```

```python
# O(n) time and space complexity

from collections import deque
def minDepth(root):
    # use level order traversal and find the first level with a leaf node
    # leaf node == no left or right child nodes
    if not root: return 0
    
    depth = 1
    q = deque([root])
    
    while q is not None:
        for _ in range(len(q)):
            node = q.popleft()
            
            if (not node.right) and (not node.left):
                return depth
            
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
                
        depth += 1
    return depth
    
    
```
