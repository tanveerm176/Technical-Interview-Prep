```python
# input: root of binary tree
# output: array of right nodes
# since we need to append the output array with right most nodes 
# we should traverse in level order using bfs

from collections import deque
def solution(root):
    # if empty tree return empty array
    if not root: return []
    
    # establsh ouput array and queue for bfs
    output = []
    queue = deque([root])
    
    # while queue is not empty
    while queue:
        
        # add last node from queue (right most to output) to the output array
        output.append(queue[-1].val)
        # get the number of nodes in the current level
        level_count = len(queue)
        
        # iterate over these nodes, remove them, and add their children if they exist
        # popleft() == remove from array and assign to var
        for _ in range(level_count):
            node = queue.popleft()
            if node.left: queue.append(node.left)
            if node.right: queue.append(node.right)

    return output
    
    
    
```
