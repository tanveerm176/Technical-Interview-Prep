```python
# Starter Code
class Node:
    def __init__(self, val, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
    
def list_to_bin_tree(l):
        if not l: return None
        
        root = Node(l.pop(0))
        queue = [root]
        
        while queue:
            node = queue.pop(0)
            
            if l:
                left_val = l.pop(0)
                if left_val != None:
                    node.left = Node(left_val)
                    queue.append(node.left)
                    
            if l:
                right_val = l.pop(0)
                if right_val != None:
                    node.right = Node(right_val)
                    queue.append(node.right)
                    
        return root
    
def bin_tree_to_list(root):
        l = []
        if not root: return l
        queue = [root]
        
        while queue:
            node = queue.pop(0)
            l.append(node.val)
            if node.left: queue.append(node.left)
            if node.right: queue.append(node.right)
            
        return l
    
```

```python
# Time: O(n) - need to traverse all nodes
# Space: O(1) - since only one var created 
#   + Best Case (balanced Tree) O(logn)
#   or + Worst Case (unbalanced) O(n) 
# Reminder we are using stack space for recursion

def invert_bin_tree(root):
    #recursive solution
    if not(root): return root
    
    temp = root.left
    root.left = invert_bin_tree(root.right)
    root.right = invert_bin_tree(temp)
    
    return root
```

```python
# Test Code
def main():
    tests = [
        {'input':[],    'output':[]},
        {'input':[2,1,3],    'output':[2,3,1]},
        {'input':[4,2,7,1,3,6,9],    'output':[4,7,2,9,6,3,1]}
    ]
    
    for i in range(len(tests)):
        print('Test' , i+1, 'Pass:',
              bin_tree_to_list(
                  invert_bin_tree(
                      list_to_bin_tree(tests[i]['input'])
                  )
              ) == tests[i]['output']
        )
        
main()
```

### Max Depth of Binary Tree

```python
def max_depth(root):
    if not(root): return 0
    
    return max(max_depth(root.left), max_depth(root.right)) + 1
```

### Min Depth of Binary Tree

```python
def min_depth(root):
    if not(root): return 0
    
    if root.left and root.right:
        return min(min_depth(root.left), min_depth(root.right)) + 1
    elif root.right:
        return min_depth(root.right) + 1
    elif root.left:
        return min_depth(root.left) + 1
    
    else:
        return 1
```

### Sorted Array to Binary Search Tree

```python
def sortedArrayToBST(nums):
        if not nums:
            return None
        
        mid_index = len(nums) // 2
        root = Node(nums[mid_index])

        root.left = sortedArrayToBST(nums[:mid_index])
        root.right = sortedArrayToBST(nums[mid_index+1:])

        return root
```
