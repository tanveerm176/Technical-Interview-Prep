```python
def solution(image):
    for row in image:
        left = 0
        right = len(row) - 1
        
        while left <= right:
            # check if there is a middle element invert that
            if left == right:
                row[left] = 1 - row[left]

            else:
                # invert the left and right elements
                temp = row[left]
                row[left] = 1 - row[right]
                row[right] = 1 - temp
                
                # can also do the following:
                """ row[left], row[right] = 1-row[right], 1-row[left] """

            left += 1
            right -= 1

    return image
```

```python
# Using boolean logic then casting
def solution(image):
    for row in image:
        left = 0
        right = len(row) - 1

        while left < right:
            temp = row[left]
            row[left] = int(not(row[right]))
            row[right] = int(not(temp))

            left += 1
            right -= 1

        if left == right:
            row[left] = int(not(row[right]))

        return image
```

```python
def main():
    tests = [
        {"input": [], "output": []},
        {"input": [[1]], "output": [[1]]},
        {"input": [[1, 2]], "output": [[1], [2]]},
        {"input": [[1, 2], [3, 4]], "output": [[1, 3], [2, 4]]},
        {
            "input": [[1, 2, 3], [4, 5, 6]],
            "output": [[1, 4], [2, 5], [3,6]],
        },
        {"input": [[1, 2, 3], [4, 5, 6]], "output": [[1, 4], [2, 5], [3, 6]]},
    ]

    for i in range(len(tests)):
        print(f"Test {i+1} Pass: {solution(tests[i]['input']) == tests[i]['output']}")


main()
```
