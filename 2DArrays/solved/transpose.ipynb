```python
def solution_neet(matrix):
    result = []
    ROWS, COLS = len(matrix), len(matrix[0])
    
    for _ in range(COLS):
        output_row = [0] * ROWS
        result.append(output_row)
    
    for r in range(ROWS):
        for c in range(COLS):
            result[c][r] = matrix[r][c]   
    
    return result

solution_neet([[1, 2, 3], [4, 5, 6]])
```

```python
#  Time: O(n*m)
#  Space: O(n*m)

'''
Since a transpose converts each column of the input into a row in the output
We need to traverse the matrix column first, 
adding the number of columns in the input as rows in the output

Then traverse row wise (going down a column) taking each element and adding it to the
row in the result array 
'''

""" 
for col in input:
    add a row in the output
    for row in input:
        add corresponding element to col in output
"""

def solution(input_matrix):
    # input: 2d matrix n*m
    # output: 2d matrix m*n
    output_matrix = []
    
    # if empty return empty
    if not input_matrix: return output_matrix
    
    input_rows = len(input_matrix)
    input_cols = len(input_matrix[0])
    
    for col in range(input_cols):
        # create an output_row
        output_row = []
        for row in range(input_rows):
            # append the elements from input to output rows
            output_row.append(input_matrix[row][col])
        
        # append the output row to the output matrix
        output_matrix.append(output_row)
    
    return output_matrix
```

### Explanation:
    1. Create an output matrix, and assign the number of rows and cols of the input to vars
    2. For each column of the input 
        2a. create a empty row for the output
        3. for each row of the input
            3a. append the element at that row,col location to the output row
        2b. append that populated row to the output matrix
    
    this is done in a for loop for every column in the input matrix

    return

```python
def main():
    tests = [
        {"input": [], "output": []},
        {"input": [[1]], "output": [[1]]},
        {"input": [[1, 2]], "output": [[1], [2]]},
        {"input": [[1, 2], [3, 4]], "output": [[1, 3], [2, 4]]},
        {
            "input": [[1, 2, 3], [4, 5, 6]],
            "output": [[1, 4], [2, 5], [3,6]],
        },
        {"input": [[1, 2, 3], [4, 5, 6]], "output": [[1, 4], [2, 5], [3, 6]]},
    ]

    for i in range(len(tests)):
        print(f"Test {i+1} Pass: {solution(tests[i]['input']) == tests[i]['output']}")


main()
```
