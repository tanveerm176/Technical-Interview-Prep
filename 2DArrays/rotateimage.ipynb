**Better Solution**

- Rotating an image is the same as transpose the image and then reversing the image
- Since this is a square image we dont need to worry about mismatched col or rows


```python
def solution(image):
    # transpose image
    for row in range(len(image)):
        for col in range(row, len(image)):
            temp = image[row][col]
            image[row][col] = image[col][row]
            image[col][row] = temp
    
        
    for row in image:
        row.reverse()
```
