### Word Pattern (Easy) [TIP103-1-1]

### Understand
    Bijection: one to one match between word in s and char in pattern

    return true or false if the pattern matches the string

1. Will there always be whitespace in the string seperating the words? YES
2. Will the words and chars always be in lower case? YES

### Match
- Can we use HashMaps/Dicts?
- How do we store key value pairs?


### Plan 
    1. Create two dicts:
        a. Map char to word
        b. Map word to char

    2. For each char/word pair in the strings:
        a. check if there is a mapping from char to word and word to char
        b. if no mapping add them
        c. if at a point there is a different mapping return False

    3. If able to match all pairs return True

### Implement:

```python
def wordPattern(pattern: str, s: str) -> bool:
    # Create a list of words from the string s
    words = s.split(" ")
    
    # If the length of the pattern is not equal to the length of the words list, return False
    if len(pattern) != len(words):
        return False
    
    # Initialize dictionaries for mapping word to character and character to word
    wordToChar = {}
    charToWord = {}
    
    # Iterate through the characters in pattern and words simultaneously
    for char, word in zip(pattern, words):
        # If the character is already in charToWord but does not map to the current word, return False
        if char in charToWord.keys() and charToWord[char] != word:
            return False
        # If the word is already in wordToChar but does not map to the current character, return False
        if word in wordToChar.keys() and wordToChar[word] != char:
            return False
        
        # Map the current word to the current character and vice versa
        wordToChar[word] = char
        charToWord[char] = word
    
    # If all checks pass, return True
    return True

```

```python
# Test cases
print(f'Test Case Passed: {wordPattern("abba", "dog cat cat dog") == True}')  # Expected output: True
print(f'Test Case Passed: {wordPattern("abba", "dog cat cat fish") == False}')  # Expected output: False
print(f'Test Case Passed: {wordPattern("aaaa", "dog dog dog dog") == True}')  # Expected output: True
print(f'Test Case Passed: {wordPattern("abba", "dog dog dog dog")== False}')  # Expected output: False
print(wordPattern("abc", "dog cat fish") == True)  # Expected output: True
print(wordPattern("abc", "dog cat cat") == False)  # Expected output: False
print(wordPattern("a", "dog") == True)  # Expected output: True
print(wordPattern("a", "dog dog") == False)  # Expected output: False
print(wordPattern("ab", "dog") == False)  # Expected output: False
```
