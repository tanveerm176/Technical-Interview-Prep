### Understand
- Prefix: a substring from the beginning of the string but before the end of the string

### Implement: 
1. Naive Approach
- start with the first letter of the first word, and check if that letter exists in every other word
-   if so add to your return string
- continue with the rest of the letters in the string and return when a char is not part of every other word

        Time Complexity

        The time complexity of the longest_common_prefix function is O(N⋅M)O(N⋅M), where:

            NN is the number of strings in the list strs.
            MM is the length of the shortest string in the list.

        Explanation:

            The outer loop iterates up to MM times (where MM is the length of the shortest string).
            The inner loop iterates N−1N−1 times (over the remaining N−1N−1 strings).

        In the worst case, every character of each string needs to be compared, leading to a total of O(N⋅M)O(N⋅M) comparisons.
        Space Complexity

        The space complexity of the function is O(1)O(1) in terms of auxiliary space, as no additional data structures are used that scale with the input size.

        Explanation:

            The prefix string itself can be considered as taking up O(M)O(M) space in the worst case where MM is the length of the shortest string, but this is the only additional space used.
            The function uses a fixed amount of extra space for variables, regardless of the input size.

        Thus, the overall space complexity is O(1)O(1), excluding the input and output.

```python
def naive_approach(strs):
    if not strs:
        return ""

    # Initialize prefix string to ""
    prefix = ""

    # Iterate over the indices i of strs[0]
    for i in range(len(strs[0])):
        # Iterate over every other string s
        for s in strs[1:]:
            # If strs[0][i] != s[i], return prefix
            if i >= len(s) or strs[0][i] != s[i]:
                return prefix

        # If all strings matched the ith character, add it to prefix
        prefix += strs[0][i]

    # Return prefix
    return prefix

# Example usage:
strs = ["flower", "flow", "flight"]
print(naive_approach(strs))  # Output: "fl"
```

### Implement 
2. Optimized Solution
- Sort all the strings in the input list
- Compare the first str and last str characters until they dont match
- return the longest prefix



```python
def longest_common_prefix(strs) -> str:
    # If the input list is empty, return an empty string
    if len(strs) == 0:
        return ''
    
    # Find the alphabetical first and last strings in the list
    first = min(strs)
    last = max(strs)
    
    # Determine the length of the shorter string between 'first' and 'last'
    length = min(len(first), len(last))
    
    # Iterate over the characters up to the length of the shorter string
    for i in range(length):
        # If characters at the same position differ, return the common prefix found so far
        if first[i] != last[i]:
            return first[0:i]
    
    # If no mismatch was found, return the entire length of the shorter string as the prefix
    return first[0:length]
```

Time Complexity

The time complexity of this solution is O(N⋅log⁡N+M)O(N⋅logN+M), where:

    NN is the number of strings in the list.
    MM is the length of the shortest string in the list.

Explanation:

    Finding the min and max strings involves comparing each string, which takes O(N⋅log⁡N)O(N⋅logN) time because it involves a sorting-like operation.
    Comparing the characters of the first and last strings takes O(M)O(M) time, where MM is the length of the shortest string.

Thus, the overall time complexity is O(N⋅log⁡N+M)O(N⋅logN+M).
Space Complexity

The space complexity of the function is O(1)O(1) in terms of auxiliary space, as no additional data structures that scale with the input size are used, aside from a few variables.

Explanation:

    The variables first, last, and length use a constant amount of space.
    The function uses a fixed amount of extra space for these variables, regardless of the input size.
