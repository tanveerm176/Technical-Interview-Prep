### Longest Substring Without Repeating Characters (Medium) [TIP103-2-2]
- LeetCode: https://leetcode.com/problems/longest-substring-without-repeating-characters/description/
- CodePath Solution:



### Understand
* How long can the strings be?
* What chars can be in the string?
- English letters, digits, symbols, and spaces
* What to return for the empty string? 0



```python
def longestSubstring(s):
    maxLength = 0
    letters = set()
    
    #init start and end at index 0 of string s
    start = 0
    for end in range(len(s)):
        #if s[end] is already in the set, advance the start pointer
        while s[end] in letters.keys():
            letters.remove(s[start])
            start += 1  
        
        letters[s[start]] = True
        
     
    
    return maxLength
```
