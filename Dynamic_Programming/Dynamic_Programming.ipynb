### Dynamic Programming
* Optimization technique applied to problems 


#### Example:
* Write a algo to find the nth Fib number 





```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    return fib(n-1) + fib(n-2)
```

```python
print(fib(4))
print(fib(5))
print(fib(8))
```

##### Works but runtime is non optimized, does a calculation for each fib number until 0


### Memoization to speed up time complexity

```python
def fib(n):
    memo = {}
    
    def fib_helper(n):
        if n in memo:
            return memo[n]
        if n == 0:
            return 0
        if n == 1:
            return 1
    
        memo[n] = fib_helper(n-1) + fib_helper(n-2)
        return memo[n]
    
    return fib_helper(n)
```
