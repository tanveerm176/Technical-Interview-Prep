#### Plan/PsuedoCode
    1. Lets use a stack to keep track of chars while parsing through the string
        a. Add to stack if it is a opening char '({['
        b. pop from stack if is a close char ')}]'
            i. This means our stack will only have opening chars
        c. if we try and pop a bracket and it is not the expected opening bracket we return False
  

##### PseudoCode

```
function valid_parens(s):
    bracket_dictionary = {
        closing_bracket : opening_bracket
    }

    opening_brackets_stack = []
    
    for char in s:
        if char in opening_bracket:
            push to stack
            skip rest of loop and continue processing string
        
        if char closing partner of top of stack:
            pop() from top of stack 
        else:
            return False

    if stack is depleted:
        return True
    else:
        return False

```

````python
def valid_parens(s):
    bracket_mapping={
        ')':'(',
        '}':'{',
        ']':'['
    }
    
    opening_bracket_stack = []
    
    for char in s:
        #if the char is not a closing bracket (must be an opening bracket) add it to the stack
        if char not in bracket_mapping.keys():
            opening_bracket_stack.append(char)
            continue
        
        #if stack exists and the top of the stack is equal to the matching char in the mapping pop from stack
        if opening_bracket_stack and (opening_bracket_stack[-1] == bracket_mapping[char]):
            opening_bracket_stack.pop()
        else:
            return False
    
    return not(opening_bracket_stack)
````

````python
print(valid_parens('{}'))
print(valid_parens('{[]'))
print(valid_parens('{]'))
print(valid_parens('{{[()]}}'))
````

#### Complexity Analysis: 
    1. Time: O(n)
        Iterate through every char 
        Adding and Popping are constant time
    2. Space: O(n)
        Using a list as a stack: 
        Worst Case: n elements used all opening brackets
        Best Case: n/2 elements used balanced string
